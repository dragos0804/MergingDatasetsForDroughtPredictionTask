```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from drought_prediction_system import DroughtPredictionSystem
```

```python
# Load your data
df = pd.read_csv("../output/AUGUMENTED_SOTWIS_NASA.csv")

# Create and train the model
model = DroughtPredictionSystem()
history = model.train(df, validation_split=0.2)
```

```python
# Plot training history
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend()
```

```python
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss (Log Scale)')
plt.ylabel('Log Loss')
plt.xlabel('Epoch')
plt.yscale('log')
plt.legend()
plt.tight_layout()
plt.show()
```
